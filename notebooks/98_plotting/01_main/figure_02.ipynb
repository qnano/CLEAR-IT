# CLEAR-IT Figure 2

Here, we generate the plots shown in Figure 2.

First, let's load the required modules.

```python
import pandas as pd
from pathlib import Path
from clearit.config import OUTPUTS_DIR
from clearit.plotting.kpplot import survival_kpplot
```

Next, we generate the Kaplan-Meier survival plots using the `survival_kpplot` function for the following cases:
- 2a, top: TNBC1-MxIF8, Discovery
- 2a, bottom: TNBC2-MIBI8, Validation
- 2b, top: TNBC2-MIBI8, Discovery
- 2b, bottom: TNBC1-MxIF8, Validation

```python
plot = survival_kpplot(
    Path(OUTPUTS_DIR,"survival_analysis","MxIF_classifier_MxIF.xls"),
    figsize=(3.1,2.6),
    title="TNBC1-MxIF8, Discovery",
    fontsize=8,
    xticks=[0,2,4,6,8,10,12,14],
    time_unit="months"
)
plot.savefig("./2a_0t.svg")
```

```python
plot = survival_kpplot(
    Path(OUTPUTS_DIR,"survival_analysis","MxIF_classifier_MIBI.xls"),
    figsize=(3.1,2.6),
    title="TNBC2-MIBI8, Validation",
    xticks=[0,2,4,6,8,10,12,14],
    fontsize=8
)
plot.savefig("./2a_1b.svg")
```

```python
plot = survival_kpplot(
    Path(OUTPUTS_DIR,"survival_analysis","MIBI_classifier_MIBI.xls"),
    figsize=(3.1,2.6),
    title="TNBC2-MIBI8, Discovery",
    xticks=[0,2,4,6,8,10,12,14],
    fontsize=8
)
plot.savefig("./2b_0t.svg")
```

```python
plot = survival_kpplot(
    Path(OUTPUTS_DIR,"survival_analysis","MIBI_classifier_MxIF.xls"),
    figsize=(3.1,2.6),
    title="TNBC1-MxIF8, Validation",
    fontsize=8,
    xticks=[0,2,4,6,8,10,12,14],
    time_unit="months"
)
plot.savefig("./2b_1b.svg")
```
